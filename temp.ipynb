```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
books = pd.read_csv("books.csv")
```

```python
books
```

```python
ax = plt.axes()
sns.heatmap(books.isna().transpose(),cbar=False,ax=ax)

plt.xlabel("Columns")
plt.ylabel("Missing Values")

plt.show()
```

```python
books["missing_description"]=np.where(books["description"].isna(),1,0)
books["age_of_book"]=2025-books["published_year"]
```

```python
col=["num_pages","age_of_book","missing_description","average_rating"]

corr=books [col].corr(method="spearman")

sns.set_theme(style="white")
plt.figure(figsize = (8,6))
heatmap= sns.heatmap(corr,annot=True,fmt=".2f",cmap="coolwarm",cbar_kws={"label":"Correlation"})

heatmap.set_title("Correlation Heatmap")
plt.show()

```

```python
book_missing=books[(~books["description"].isna()) & ~(books["num_pages"].isna()) & ~ (books["average_rating"].isna()) & ~ (books["published_year"].isna())]
```

```python
book_missing
```

```python
book_missing["categories"].value_counts().sort_values(ascending=False).reset_index()
```

```python
book_missing["words_in_description"]=book_missing["description"].str.split().str.len()
```

```python
book_missing_25words=book_missing[book_missing["words_in_description"]>=25]
book_missing_25words
```

```python
book_missing_25words["title_and_subtitle"]=(np.where(book_missing_25words["subtitle"].isna(),book_missing_25words["title"],book_missing_25words[["title","subtitle"]].astype(str).agg(": ".join,axis=1)))
```

```python
book_missing_25words=book_missing_25words.copy()
book_missing_25words
```

```python
book_missing_25words["tagged_description"]=book_missing_25words[["isbn13","description"]].astype(str).agg(" ".join,axis=1)
```

```python
book_missing_25words.drop(["subtitle","missing_description","age_of_book","words_in_description"],axis=1).to_csv("book_cleaned.csv",index=False)
```
